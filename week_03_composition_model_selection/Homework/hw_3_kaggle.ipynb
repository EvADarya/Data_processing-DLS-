```python
data = pd.read_csv('./train.csv')
```

```python
#        

#  
num_cols = [
    'ClientPeriod',
    'MonthlySpending',
    'TotalSpent'
]

#  
cat_cols = [
    'Sex',
    'IsSeniorCitizen',
    'HasPartner',
    'HasChild',
    'HasPhoneService',
    'HasMultiplePhoneNumbers',
    'HasInternetService',
    'HasOnlineSecurityService',
    'HasOnlineBackup',
    'HasDeviceProtection',
    'HasTechSupportAccess',
    'HasOnlineTV',
    'HasMovieSubscription',
    'HasContractPhone',
    'IsBillingPaperless',
    'PaymentMethod'
]

feature_cols = num_cols + cat_cols
target_col = 'Churn'
```

```python
num_data = data[num_cols]
num_data
```

```python
cat_data = data[cat_cols].copy()
for i in cat_cols:
  print(f"{i} : {data[i].unique()}")
```

```python
cat_data['HasChild'] = np.where(data['HasChild'] == "Yes", 1, 0)
cat_data['Sex'] = np.where(data['Sex'] == "Male", 1, 0)
cat_data['HasPartner'] = np.where(data['HasPhoneService'] == "Yes", 1, 0)
cat_data['HasPhoneService'] = np.where(data['HasPhoneService'] == "Yes", 1, 0)
cat_data['IsBillingPaperless'] = np.where(data['IsBillingPaperless'] == "Yes", 1, 0)
```

```python
dummies_cols = ['HasMultiplePhoneNumbers',
                'HasInternetService',
                'HasOnlineSecurityService',
                'HasOnlineBackup',
                'HasDeviceProtection',
                'HasTechSupportAccess',
                'HasOnlineTV',
                'HasMovieSubscription',
                'HasContractPhone']
dummies_data = pd.get_dummies(data[dummies_cols],drop_first= True, dtype = int)
```

```python
cat_data = pd.concat([dummies_data, cat_data.drop(dummies_cols, axis =1)], axis = 1)
```

```python
encoder = LabelEncoder()
cat_data['PaymentMethod'] = encoder.fit_transform(cat_data['PaymentMethod'])
```

```python
cat_data.info()
```

```python
new_data = pd.concat([cat_data, num_data],axis = 1)
```

```python
from math import nan
new_data['TotalSpent'].value_counts()
new_data['TotalSpent'] = (new_data['TotalSpent'].replace(" ", nan)).astype(float)
```

```python
new_data['TotalSpent'] = new_data['TotalSpent'].fillna(new_data['TotalSpent'].mean())
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
X_train,X_test, y_train,y_test = train_test_split(new_data, data['Churn'],test_size = 0.2,random_state = 42,shuffle = True)
model = LogisticRegression(random_state = 42)

param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}

grid_search = GridSearchCV(
    model,
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

print(f"Best params: {grid_search.best_params_}")
print(f"Best score: {grid_search.best_score_}")
best_model = grid_search.best_estimator_
```

```python
from sklearn.metrics import roc_auc_score
```

```python
y_test_pred = best_model.predict_proba(X_test)[:, 1]

test_score = roc_auc_score(y_test, y_test_pred)
print(f"Test score:    {test_score:.4f}")
```

```python
import catboost
```

```python
boosting_model = catboost.CatBoostClassifier(n_estimators=200, logging_level='Silent',
                                             eval_metric='AUC',random_state= 42)

X_train_b,X_test_b, y_train_b,y_test_b = train_test_split(new_data, data['Churn'],test_size = 0.2,random_state = 42,shuffle = True)

param_grid_catboost = {'depth': [4, 6, 8, 10],
    'l2_leaf_reg': [1, 3, 5, 7, 9],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 0.9, 1.0] }
boosting_model.grid_search(param_grid_catboost,
                           X_train_b,
                           y_train_b, refit=True, verbose=False);
boosting_model.fit(X_train_b, y_train_b)

y_train_b_predicted = boosting_model.predict_proba(X_train_b)[:, 1]
y_test__b_predicted = boosting_model.predict_proba(X_test_b)[:, 1]

# ROC-AUC scores
train_auc = roc_auc_score(y_train_b, y_train_b_predicted)
test_auc = roc_auc_score(y_test_b, y_test__b_predicted)

print(f"Train ROC-AUC: {train_auc:.4f}")
print(f"Test ROC-AUC:  {test_auc:.4f}")
```
